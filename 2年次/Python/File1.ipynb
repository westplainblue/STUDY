```python
from sklearn import datasets
from sklearn import svm
```

```python
iris=datasets.load_iris()
print(iris.data)
print(iris.data.shape)
```

```python
num = len(iris.data)
print(num)
```

```python
clf = svm.SVC(gamma="auto")
clf.fit(iris.data, iris.target)
```

```python
print(clf.predict([[6.9, 3.1, 5.4, 2.1]]))#左からがくの長さ、がくの幅、花びらの長さ、花びらの幅を入力する入力する
#結果が０だとSetosa、１だとVersicolor、２だとVersinicaという結果になる
#これらのプログラミングは教師あり学習のクラス分類になる。
#参考文献https://aiacademy.jp/media/?p=1916
```
